#### for plotting heatmap we need to convert all categorical columns in numericals

```python
# checking value counts of column 'gender' 

df_1['Gender'].value_counts()
```

```python
# replacing value with numerics

df_1['Gender'] = df_1['Gender'].replace({'Male':0,'Female':1})
```

```python
# checking value counts of column 'Senior Management'

df_1['Senior Management'].value_counts()
```

```python
# replacing value with numerics

df_1['Senior Management'] = df_1['Senior Management'].replace({True:0,False:1})
```

```python
# checking value counts of column 'team'

df_1['Team'].value_counts()
```

```python
# replacing value with numerics

df_1['Team'] = df_1['Team'].replace({'Marketing':1,'Finance':2,'Business Development':3,'Client Services':4,'Product':5,'Engineering':6,
                                     'Human Resources':7,'Sales':8,'Legal':9,'Distribution':10})
```

```python
# Extracting year from date column 

df_1['Start Date'] = pd.to_datetime(df_1['Start Date'])

df_1['year'] = df_1['Start Date'].dt.year
```

```python
# after changes checking the data 

df_1.head()
```

```python
# dropping categorical column (first name , start date , last login time) as it does not useful for plotting heatmap 

df_1 = df_1.drop(['First Name','Start Date','Last Login Time'], axis=1)
```

```python
# after dopping just cross checking 

df_1.head()
```

```python
# plotting a heatmap 

sns.heatmap(df_1.corr(),annot=True,cmap='coolwarm')
plt.show()
```
