```python
# for finding correlation between those columns,now making a dataframe from those columns

corr_cols = df[['num_critic_for_reviews','duration','director_facebook_likes','actor_3_facebook_likes','actor_1_facebook_likes','gross','facenumber_in_poster','num_user_for_reviews','budget','title_year','actor_2_facebook_likes']]
```
